#STATISTICAL DATA ANALYSIS ADVANCED
* MAPPING JOBS TO CATEGORICAL DATA
* IMPORTING NECCESARY SCIPY.STATS

```python
import scipy.stats as stats

job_mappings={
    'CHP Helper' : 1 , 'Sub-station Operator' : 2, 'Helper' : 3, 'Rigor Tandal' : 4,
       'Technician' : 5, 'Maintenance Rigor' : 6, 'Mechanical Helper' : 7,'Instrument Helper' : 8
}
df["Job Role"].replace(job_mappings, inplace=True)
f, p = stats.f_oneway(df['I receive clear instructions and guidance from my supervisors.'], df['I feel comfortable approaching my supervisors with questions or concerns.'])
print(f'ANOVA Result: F={f:.2f}, p={p:.3f}')

if p < 0.05:
    print('Significant difference in means between supervisors guidance')
else:
    print('No significant difference in means between supervisors guidance')
```

# ANNOVA TEST OF RESPONSES




```python
f, p = stats.f_oneway(df['I try to acquire new knowledge and new skills through self learning and peer learning.'], df['I feel comfortable approaching my supervisors with questions or concerns.'])
print(f'ANOVA Result: F={f:.2f}, p={p:.3f}')

if p < 0.05:
    print('Significant difference in means between supervisors guidance and self learning')
else:
    print('No significant difference in means between supervisors guidance and self learning')
```

# ANNOVA TEST CONTINUED

```python
# print(col)
f, p = stats.f_oneway(df[col[12]], df[col[15]])
print(f'ANOVA Result: F={f:.2f}, p={p:.3f}')

if p < 0.05:
    print('Significant difference in means between peer learning and discrimnation')
else:
    print('No significant difference in means between peer learning and discrimnation')
```

# F - TEST AND T - TEST
